```python
import pandas as pd
from IPython.display import display, HTML

# Функція для відображення DataFrame з прокруткою
def display_scrollable_dataframe(df, max_height=300):
    display(HTML(f"""
        <div style="height:{max_height}px; overflow:auto; border:1px solid black;">
            {df.to_html(index=False)}
        </div>
    """))

raw_data = {
    'KATOTG_lvl5': ['Київ', 'Київ', 'Ірпінь', 'Львів', 'Харків', 'Одеса'],
    'KATOTG_lvl1': ['Київ', 'Київ', 'Київ', 'Львів', 'Харків', 'Одеса'],
    'p_lvl2': ['p1-1', 'p1-2', 'p1-2', 'p2-1', 'p2-2', 'p2-3'],
    'p_lvl1': ['p1', 'p1', 'p1', 'p2', 'p2', 'p2'],
    'v': [50, 50, 3, 34, 29, 25],
}

df = pd.DataFrame(raw_data)
df['_T'] = '_T'

print(df)
```

```python
KATOTG_lvl_list = ['KATOTG_lvl5','KATOTG_lvl1', '_T']
p_lvl_list = ['p_lvl1', 'p_lvl2', '_T']

grouped_column = ['KATOTG', 'Par']
all_column = grouped_column + ['OBS']
result_df = pd.DataFrame()

for column1 in KATOTG_lvl_list:
    for column2 in p_lvl_list:
        tmp_column = [column1, column2]
        tmp_df = df[tmp_column+['v']]
        tmp_df.columns = all_column
        grouped_df = tmp_df.groupby(grouped_column)['OBS'].sum().reset_index()
        result_df = pd.concat([result_df, grouped_df], ignore_index=True)

# Виклик функції для відображення DataFrame з прокруткою
display_scrollable_dataframe(result_df)
```
